1.2	Find the cuisines which are not present in restaurant of Delhi NCR but present in rest of India.

```python
data_ncr=data_india[data_india['Delhi_Rest_of_India']=='Delhi NCR']
data_rest=data_india[data_india['Delhi_Rest_of_India']=="Rest of India"]
```

```python
split_cuisines_delhi_ncr=data_ncr['Cuisines'].str.split(',',expand=True)


split_cuisines_rest=data_rest['Cuisines'].str.split(',',expand=True)
```

```python
# Creating a two lists of all cuisines available in delhi NCR and Rest of India

all_cuisines_delhi_ncr=[]
for c in range(split_cuisines_delhi_ncr.shape[1]):
    for r in range(split_cuisines_delhi_ncr.shape[0]):
        if split_cuisines_delhi_ncr.iloc[r,c] not in all_cuisines_delhi_ncr and split_cuisines_delhi_ncr.iloc[r,c] is not None:
            all_cuisines_delhi_ncr.append(split_cuisines_delhi_ncr.iloc[r,c].lstrip())
        
        
all_cuisines_rest=[]
for c in range(split_cuisines_rest.shape[1]):
    for r in range(split_cuisines_rest.shape[0]):
        if split_cuisines_rest.iloc[r,c] not in all_cuisines_rest and split_cuisines_rest.iloc[r,c] is not None:
            all_cuisines_rest.append(split_cuisines_rest.iloc[r,c].lstrip())
```

```python
all_cuisines_delhi_ncr
```

```python
data_ncr.shape,data_rest.shape
```

```python
results=[]

for i in all_cuisines_rest:
    if i not in all_cuisines_delhi_ncr:
        results.append(i)
```

```python
len(all_cuisines_delhi_ncr),len(all_cuisines_rest),len(results)
```

2.2.	Find the weighted restaurant rating of each locality and find out the top 10 localities with more weighted restaurant rating?


**Weighted Restaurant Rating=Σ (number of votes * rating) / Σ (number of votes) .**


```python
data['Rating X Votes']=data['Aggregate rating']*data['Votes']
```

```python
grp_data_vote=data.groupby("Locality").sum()[['Rating X Votes','Votes']]
```

```python
grp_data_vote['Weighted_Rating']=grp_data_vote['Rating X Votes']/grp_data_vote['Votes']
grp_data_vote.sort_values('Weighted_Rating',ascending=False)[:10]
```

3.5.	Plot the bubble graph of a number of Restaurants present in the city of India and keeping the weighted restaurant rating of the city in a bubble.

```python
data_india=data[data['Country Code']==1]
grp_data_vote2=data_india.groupby("City",as_index=False).sum()[['City','Rating X Votes','Votes']]
grp_data_vote2['Weighted Mean']=np.round(grp_data_vote2['Rating X Votes']/grp_data_vote2['Votes'],2)
grp_data_vote2['Count_of_Restaurants']=data_india.groupby("City").count()['Votes'].to_list()

grp_data_vote2
```

```python
data_india.groupby("City").count()['Votes'].shape
```

```python
import plotly.express as px
df = px.data.gapminder()

fig = px.scatter(grp_data_vote2, x="City",
	         size="Count_of_Restaurants", y="Weighted Mean",
            color="Weighted Mean",size_max=150,hover_data=['Weighted Mean'],
                 title="Number of Restaurants and Weighted Ratings")

fig.show()
```
